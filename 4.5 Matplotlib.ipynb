### Exercise 1
Nicholas and Jordan went on holidays with their families. They decided to monitor the kilometers they traveled by keeping track of the time and the distance they were on the road. The boys collected the following data:  

Nicholas:  

Time (hr)| 1 | 2 | 3 | 4 | 5 | 6
--------|---|---|---|---|---|---
 **Distance (kilometers)**| 60 | 110 | 175 | 235 | 280 | 320


Jordan:
Time (hr)| 1 | 2 | 3 | 4 | 5 | 6
--------|---|---|---|---|---|---
 **Distance (kilometers)**| 50 | 90 | 125 | 125 | 165 | 210

* Draw a graph to show the trip of each boy.  
* What conclusions could you draw by looking at the graphs?

![](files/2023-11-16-08-42-06.png)

````python
import numpy as np 

nicholas = np.array([0,60,110,175,235,280,320])
jordan = np.array([0,50,90,125,125,165,210])
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Data arrays (including 0,0 as starting point)
nicholas = np.array([0, 60, 110, 175, 235, 280, 320])
jordan = np.array([0, 50, 90, 125, 125, 165, 210])
time = np.array([0, 1, 2, 3, 4, 5, 6])

# Create the graph
plt.figure(figsize=(10, 6))

# Plot both lines
plt.plot(time, nicholas, 'b-o', linewidth=2, markersize=8, label='Nicholas')
plt.plot(time, jordan, 'r-s', linewidth=2, markersize=8, label='Jordan')

# Add labels and title
plt.xlabel('Time (hours)', fontsize=12)
plt.ylabel('Distance (kilometers)', fontsize=12)
plt.title('Distance Traveled Over Time', fontsize=14, fontweight='bold')

# Add grid for better readability
plt.grid(True, alpha=0.3)

# Add legend
plt.legend(fontsize=12)

# Set axis limits for better view
plt.xlim(0, 6.5)
plt.ylim(0, 350)

# Add minor grid lines
plt.minorticks_on()
plt.grid(True, which='minor', alpha=0.1)

# Show the plot
plt.tight_layout()
plt.show()

# Calculate some statistics for analysis
print("=" * 50)
print("ANALYSIS OF TRAVEL DATA")
print("=" * 50)

# Total distance
print(f"\nTotal distance traveled (6 hours):")
print(f"  Nicholas: {nicholas[-1]} km")
print(f"  Jordan: {jordan[-1]} km")

# Average speed
print(f"\nAverage speed over entire trip:")
print(f"  Nicholas: {nicholas[-1]/6:.1f} km/h")
print(f"  Jordan: {jordan[-1]/6:.1f} km/h")

# Segment speeds (distance traveled each hour)
nicholas_segments = np.diff(nicholas)
jordan_segments = np.diff(jordan)

print(f"\nDistance traveled each hour (km):")
print(f"  Hour  1   2   3   4   5   6")
print(f"  Nicholas: {nicholas_segments}")
print(f"  Jordan:   {jordan_segments}")

print(f"\nSpeed during each hour (km/h):")
print(f"  Hour     1     2     3     4     5     6")
print(f"  Nicholas: {nicholas_segments}")
print(f"  Jordan:   {jordan_segments}")

# Identify patterns
print(f"\n" + "=" * 50)
print("CONCLUSIONS")
print("=" * 50)

# Nicholas analysis
print("\n NICHOLAS:")
print("  • Consistent traveler - distance increases every hour")
print(f"  • Fastest hour: Hour {np.argmax(nicholas_segments)+1} ({np.max(nicholas_segments)} km)")
print(f"  • Slowest hour: Hour {np.argmin(nicholas_segments)+1} ({np.min(nicholas_segments)} km)")
print("  • Shows good endurance with steady progress")

# Jordan analysis
print("\n JORDAN:")
print("  • Interesting pattern - no progress in hour 4 (0 km gained)")
print(f"  • Fastest hour: Hour {np.argmax(jordan_segments)+1} ({np.max(jordan_segments)} km)")
print("  • Possible break or stop during hour 4")
print("  • Speed decreases significantly after hour 3")

# Comparison
print("\n COMPARISON:")
if nicholas[-1] > jordan[-1]:
    print(f"  • Nicholas traveled {nicholas[-1] - jordan[-1]} km further than Jordan")
else:
    print(f"  • Jordan traveled {jordan[-1] - nicholas[-1]} km further than Nicholas")

print(f"  • Nicholas's speed: {nicholas_segments.mean():.1f} km/h average")
print(f"  • Jordan's speed:   {jordan_segments.mean():.1f} km/h average")
print(f"  • Nicholas was {abs(nicholas_segments.mean() - jordan_segments.mean()):.1f} km/h {'faster' if nicholas_segments.mean() > jordan_segments.mean() else 'slower'} on average")

# Check for breaks
if 0 in jordan_segments:
    zero_hour = np.where(jordan_segments == 0)[0][0] + 1
    print(f"\n    Jordan likely took a break during hour {zero_hour} (0 km traveled)")
````

### Exercise 2
Here is the data for the number of ice cream cones sold each week at an ice cream stand during the months of June, July and August.  


Week |June |July |August
---------|----------|---------|---------
 Week 1 | 500 | 800 |700
 Week 2 | 800 | 900 |1000
 Week 3| 700 | 600 |800
 Week 4| 800 | 800 |900  

Create a bar chart to compare the weekly data for June, July and August.

![](files/2023-11-16-08-40-17.png)

````python
import numpy as np 
number_of_cones_june = np.array([500,800,700,800]) 
number_of_cones_july = np.array([800,900,600,800]) 
number_of_cones_august = np.array([700,1000,800,900])
weeks = ('week 1', 'week 2', 'week 3' , 'week 4')
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Data arrays
number_of_cones_june = np.array([500, 800, 700, 800])
number_of_cones_july = np.array([800, 900, 600, 800])
number_of_cones_august = np.array([700, 1000, 800, 900])
weeks = ('Week 1', 'Week 2', 'Week 3', 'Week 4')

# Set up the bar chart
x = np.arange(len(weeks))
width = 0.25

fig, ax = plt.subplots(figsize=(12, 7))

# Create bars for each month
bars_june = ax.bar(x - width, number_of_cones_june, width, label='June', color='lightcoral', edgecolor='black')
bars_july = ax.bar(x, number_of_cones_july, width, label='July', color='gold', edgecolor='black')
bars_august = ax.bar(x + width, number_of_cones_august, width, label='August', color='lightblue', edgecolor='black')

# Add labels, title, and legend
ax.set_xlabel('Week', fontsize=12, fontweight='bold')
ax.set_ylabel('Number of Ice Cream Cones Sold', fontsize=12, fontweight='bold')
ax.set_title('Ice Cream Cone Sales by Month and Week', fontsize=14, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(weeks)
ax.legend(fontsize=11)

# Add value labels on top of bars
for bar in bars_june:
    height = bar.get_height()
    ax.annotate(f'{height}', xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=9)
for bar in bars_july:
    height = bar.get_height()
    ax.annotate(f'{height}', xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=9)
for bar in bars_august:
    height = bar.get_height()
    ax.annotate(f'{height}', xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', fontsize=9)

# Add grid
ax.grid(True, alpha=0.3, axis='y')
ax.set_ylim(0, 1200)

# Add average lines
avg_june = np.mean(number_of_cones_june)
avg_july = np.mean(number_of_cones_july)
avg_august = np.mean(number_of_cones_august)

ax.axhline(y=avg_june, color='lightcoral', linestyle='--', alpha=0.5, linewidth=1)
ax.axhline(y=avg_july, color='gold', linestyle='--', alpha=0.5, linewidth=1)
ax.axhline(y=avg_august, color='lightblue', linestyle='--', alpha=0.5, linewidth=1)

plt.tight_layout()
plt.show()

# Print analysis
print("=" * 60)
print("ICE CREAM CONE SALES ANALYSIS")
print("=" * 60)

print("\nSALES BY WEEK:")
print("-" * 40)
print(f"{'Week':<10} {'June':<10} {'July':<10} {'August':<10}")
print("-" * 40)
for i, week in enumerate(weeks):
    print(f"{week:<10} {number_of_cones_june[i]:<10} {number_of_cones_july[i]:<10} {number_of_cones_august[i]:<10}")

print("\nTOTALS AND AVERAGES:")
print("-" * 40)
print(f"Month     Total    Average   Best Week")
print("-" * 40)
print(f"June:     {np.sum(number_of_cones_june):<8} {np.mean(number_of_cones_june):.0f}      Week {np.argmax(number_of_cones_june)+1}")
print(f"July:     {np.sum(number_of_cones_july):<8} {np.mean(number_of_cones_july):.0f}      Week {np.argmax(number_of_cones_july)+1}")
print(f"August:   {np.sum(number_of_cones_august):<8} {np.mean(number_of_cones_august):.0f}      Week {np.argmax(number_of_cones_august)+1}")

print("\nKEY OBSERVATIONS:")
print("-" * 40)

all_sales = np.concatenate([number_of_cones_june, number_of_cones_july, number_of_cones_august])
best_sale = np.max(all_sales)
worst_sale = np.min(all_sales)

print(f"• Best sales week: {best_sale} cones (August Week 2)")
print(f"• Worst sales week: {worst_sale} cones (June Week 1)")
print(f"• July has the highest average sales ({np.mean(number_of_cones_july):.0f} cones/week)")
print(f"• June has the most consistent sales (range: {np.max(number_of_cones_june)-np.min(number_of_cones_june)} cones)")
print(f"• August shows strong late-month sales (Week 4: 900 cones)")

print("\nMONTHLY COMPARISON:")
print("-" * 40)
months = ['June', 'July', 'August']
totals = [np.sum(number_of_cones_june), np.sum(number_of_cones_july), np.sum(number_of_cones_august)]
best_month = months[np.argmax(totals)]
print(f"• {best_month} had the highest total sales ({np.max(totals)} cones)")
print(f"• July outsold June by {totals[1]-totals[0]} cones")
print(f"• August outsold June by {totals[2]-totals[0]} cones")
````

### Exercise 3
Scientists have invented a new dietary supplement that is supposed to increase the weight of a piglet within its first 3 months of growth. Farmer John fed this supplement to his stock of piglets, and at the end of 3 months, he recorded the weights of 50 randomly selected piglets.

The following array is the recorded weights (in pounds) of the 50 selected piglets:  
```
_array([115, 121, 100,  80, 114,  92, 112,  65,  90, 91,  113,  82, 121,
       120, 118,  94,  96, 121, 116, 107,  89, 108, 112,  67,  85,  86,
        95, 101, 107,  98,  96,  88,  108,  112, 114, 122,  85,  105,  83,123,  97, 116, 94, 120,  80,  94, 107,  68,  110,  81])_  
```

Draw a graph to represent this data in a meaningful way.
What can you conclude?


````python
import numpy as np


piglets = np.array([115, 121, 100,  80, 114,  92, 112,  65,  90, 91,  113,  82, 121,
       120, 118,  94,  96, 121, 116, 107,  89, 108, 112,  67,  85,  86,
        95, 101, 107,  98,  96,  88,  108,  112, 114, 122,  85,  105,  83,
       123,  97, 116, 94, 120,  80,  94, 107,  68,  110,  81])
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Data
piglets = np.array([115, 121, 100,  80, 114,  92, 112,  65,  90, 91,  113,  82, 121,
       120, 118,  94,  96, 121, 116, 107,  89, 108, 112,  67,  85,  86,
        95, 101, 107,  98,  96,  88,  108,  112, 114, 122,  85,  105,  83,
       123,  97, 116, 94, 120,  80,  94, 107,  68,  110,  81])

# Create the histogram
plt.figure(figsize=(12, 6))

# Histogram with frequency
plt.subplot(1, 2, 1)
plt.hist(piglets, bins=12, edgecolor='black', alpha=0.7, color='skyblue')
plt.xlabel('Weight (pounds)', fontsize=12)
plt.ylabel('Number of Piglets', fontsize=12)
plt.title('Distribution of Piglet Weights', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

# Add mean and median lines
mean_weight = np.mean(piglets)
median_weight = np.median(piglets)
plt.axvline(mean_weight, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_weight:.1f} lbs')
plt.axvline(median_weight, color='green', linestyle='--', linewidth=2, label=f'Median: {median_weight:.1f} lbs')
plt.legend()

# Box plot for additional insights
plt.subplot(1, 2, 2)
plt.boxplot(piglets, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.ylabel('Weight (pounds)', fontsize=12)
plt.title('Box Plot of Piglet Weights', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.xticks([1], ['Piglets'])

plt.tight_layout()
plt.show()

# Statistical analysis
print("=" * 60)
print("PIGLET WEIGHT ANALYSIS")
print("=" * 60)

print(f"\nBASIC STATISTICS:")
print("-" * 40)
print(f"Number of piglets: {len(piglets)}")
print(f"Mean weight: {np.mean(piglets):.2f} lbs")
print(f"Median weight: {np.median(piglets):.2f} lbs")
print(f"Minimum weight: {np.min(piglets)} lbs")
print(f"Maximum weight: {np.max(piglets)} lbs")
print(f"Range: {np.max(piglets) - np.min(piglets)} lbs")
print(f"Standard deviation: {np.std(piglets):.2f} lbs")
print(f"Variance: {np.var(piglets):.2f}")

print(f"\nQUARTILES:")
print("-" * 40)
q1 = np.percentile(piglets, 25)
q2 = np.percentile(piglets, 50)  # same as median
q3 = np.percentile(piglets, 75)
print(f"Q1 (25th percentile): {q1:.2f} lbs")
print(f"Q2 (50th percentile/Median): {q2:.2f} lbs")
print(f"Q3 (75th percentile): {q3:.2f} lbs")
print(f"IQR (Interquartile Range): {q3 - q1:.2f} lbs")

print(f"\nDISTRIBUTION SHAPE:")
print("-" * 40)
if mean_weight > median_weight:
    print("• Slightly right-skewed (mean > median)")
elif mean_weight < median_weight:
    print("• Slightly left-skewed (mean < median)")
else:
    print("• Approximately symmetric (mean ≈ median)")

# Check for outliers using IQR method
lower_bound = q1 - 1.5 * (q3 - q1)
upper_bound = q3 + 1.5 * (q3 - q1)
outliers = piglets[(piglets < lower_bound) | (piglets > upper_bound)]
print(f"\nOUTLIERS:")
print("-" * 40)
print(f"Lower fence: {lower_bound:.2f} lbs")
print(f"Upper fence: {upper_bound:.2f} lbs")
if len(outliers) > 0:
    print(f"Outliers found: {sorted(outliers)}")
    print(f"Number of outliers: {len(outliers)}")
else:
    print("No significant outliers detected")

print(f"\nWEIGHT DISTRIBUTION:")
print("-" * 40)
# Create weight categories
bins = [60, 70, 80, 90, 100, 110, 120, 130]
labels = ['60-70', '70-80', '80-90', '90-100', '100-110', '110-120', '120-130']
counts, _ = np.histogram(piglets, bins=bins)

for i, (label, count) in enumerate(zip(labels, counts)):
    percentage = (count / len(piglets)) * 100
    print(f"{label} lbs: {count} piglets ({percentage:.1f}%)")
````

### Exercise 4

The following table consists of the marks achieved by 9 students on chemistry and math tests:


Student | A | B | C | D | E | F | G | H | I
--------|---|---|---|---|---|---|---|---|--
 **Chemistry marks** | 49 | 46 | 35 | 58 | 51 | 56 | 46 | 53 | 39
 **Math marks** | 29 | 23 | 10 | 40 | 38 | 36 | 32 | 28| 18 | 

The question is whether students with high marks in maths are also good at chemistry.  Choose a graph type suitable for investigating whether or not there a relationship or connection exists between 2 sets of data.



````python
import numpy as np
chemistry_marks = np.array([49, 46,35,58,51,56, 46, 35, 39])
math_marks = np.array([29, 23,10,40,38,36, 32, 28, 18])
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Data
chemistry_marks = np.array([49, 46, 35, 58, 51, 56, 46, 35, 39])
math_marks = np.array([29, 23, 10, 40, 38, 36, 32, 28, 18])
students = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']

# Create the scatter plot
plt.figure(figsize=(10, 8))

# Scatter plot with student labels
plt.scatter(chemistry_marks, math_marks, color='darkblue', s=100, alpha=0.7, edgecolor='black')

# Add student labels to each point
for i, student in enumerate(students):
    plt.annotate(student, (chemistry_marks[i] + 0.5, math_marks[i] + 0.5), fontsize=11, fontweight='bold')

# Add a trend line (linear regression)
z = np.polyfit(chemistry_marks, math_marks, 1)
p = np.poly1d(z)
plt.plot(chemistry_marks, p(chemistry_marks), "r--", linewidth=2, label=f'Trend line (slope = {z[0]:.2f})')

# Add labels and title
plt.xlabel('Chemistry Marks', fontsize=13, fontweight='bold')
plt.ylabel('Math Marks', fontsize=13, fontweight='bold')
plt.title('Relationship Between Chemistry and Math Marks', fontsize=15, fontweight='bold')

# Add grid
plt.grid(True, alpha=0.3)

# Set axis limits for better view
plt.xlim(30, 65)
plt.ylim(5, 45)

# Add legend
plt.legend(fontsize=11)

plt.tight_layout()
plt.show()

# Calculate correlation
print("=" * 60)
print("RELATIONSHIP BETWEEN CHEMISTRY AND MATH MARKS")
print("=" * 60)

# Pearson correlation coefficient
correlation = np.corrcoef(chemistry_marks, math_marks)[0, 1]
print(f"\nPearson correlation coefficient: {correlation:.3f}")

# Interpret correlation strength
print("\nCORRELATION STRENGTH:")
print("-" * 40)
if abs(correlation) > 0.8:
    strength = "Very strong"
elif abs(correlation) > 0.6:
    strength = "Strong"
elif abs(correlation) > 0.4:
    strength = "Moderate"
elif abs(correlation) > 0.2:
    strength = "Weak"
else:
    strength = "Very weak"

direction = "positive" if correlation > 0 else "negative"
print(f"• {strength} {direction} correlation")

# Calculate linear regression
slope, intercept = np.polyfit(chemistry_marks, math_marks, 1)
print(f"\nLINEAR REGRESSION:")
print("-" * 40)
print(f"Equation: Math = {slope:.2f} × Chemistry + {intercept:.2f}")
print(f"Slope: {slope:.2f} (for every 1 point increase in Chemistry, Math changes by {slope:.2f} points)")

# Calculate statistics for each subject
print(f"\nSUBJECT STATISTICS:")
print("-" * 40)
print(f"{'Subject':<12} {'Mean':<8} {'Std Dev':<10} {'Min':<6} {'Max':<6}")
print("-" * 40)
print(f"{'Chemistry':<12} {np.mean(chemistry_marks):<8.1f} {np.std(chemistry_marks):<10.2f} {np.min(chemistry_marks):<6} {np.max(chemistry_marks):<6}")
print(f"{'Math':<12} {np.mean(math_marks):<8.1f} {np.std(math_marks):<10.2f} {np.min(math_marks):<6} {np.max(math_marks):<6}")

# Identify specific observations
print(f"\nOBSERVATIONS:")
print("-" * 40)

# Find highest and lowest in each subject
highest_chem_idx = np.argmax(chemistry_marks)
lowest_chem_idx = np.argmin(chemistry_marks)
highest_math_idx = np.argmax(math_marks)
lowest_math_idx = np.argmin(math_marks)

print(f"• Student {students[highest_chem_idx]} has highest Chemistry mark ({chemistry_marks[highest_chem_idx]})")
print(f"  → Math mark: {math_marks[highest_chem_idx]}")
print(f"• Student {students[lowest_chem_idx]} has lowest Chemistry mark ({chemistry_marks[lowest_chem_idx]})")
print(f"  → Math mark: {math_marks[lowest_chem_idx]}")
print(f"• Student {students[highest_math_idx]} has highest Math mark ({math_marks[highest_math_idx]})")
print(f"  → Chemistry mark: {chemistry_marks[highest_math_idx]}")
print(f"• Student {students[lowest_math_idx]} has lowest Math mark ({math_marks[lowest_math_idx]})")
print(f"  → Chemistry mark: {chemistry_marks[lowest_math_idx]}")

# Conclusion
print(f"\nCONCLUSION:")
print("-" * 40)
if correlation > 0.5:
    print(f"• There is a {strength.lower()} positive relationship between Chemistry and Math marks.")
    print("• Students who perform well in Chemistry tend to also perform well in Math.")
    print("• This suggests that skills or study habits might transfer between subjects.")
elif correlation > 0.3:
    print("• There is a moderate positive relationship between the two subjects.")
    print("• Some connection exists, but it's not extremely strong.")
elif correlation > 0:
    print("• There is a weak positive relationship between Chemistry and Math marks.")
    print("• Performance in one subject is not a strong predictor of performance in the other.")
else:
    print("• There is a negative relationship (unusual for related subjects).")
    print("• Students good at Chemistry might struggle with Math, or vice versa.")

print(f"\nNote: Correlation coefficient of {correlation:.3f} indicates a "
      f"{'strong' if abs(correlation) > 0.7 else 'moderate' if abs(correlation) > 0.5 else 'weak'} "
      f"relationship.")
````

### Exercise 5

The Red Cross Blood Donor Clinic had a very successful morning collecting blood donations. Within 3 hours, 25 people had made donations, and the following is a table showing the blood types of the donations:

Blood Type | A | B | O | AB
-----------|---|---|---|---
 **Number of donors** | 9 | 5 | 7 | 4

Plot the donations so that it is clear what percentage of each blood type was collected that morning.

````python
import numpy as np
import matplotlib.pyplot as plt

# Data
blood_types = np.array(['A', 'B', 'O', 'AB'])
number_of_donors = np.array([9, 5, 7, 4])
total_donors = np.sum(number_of_donors)
percentages = (number_of_donors / total_donors) * 100

# Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Option 1: Pie Chart
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
explode = (0.05, 0.05, 0.05, 0.05)  # Slight separation

wedges, texts, autotexts = ax1.pie(number_of_donors, 
                                    labels=blood_types,
                                    autopct='%1.1f%%',
                                    startangle=90,
                                    colors=colors,
                                    explode=explode,
                                    shadow=True)

# Make percentage text bold and white
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')
    autotext.set_fontsize(12)

ax1.set_title('Blood Type Donations (Pie Chart)', fontsize=14, fontweight='bold')
ax1.axis('equal')  # Equal aspect ratio ensures pie is circular

# Option 2: Bar Chart with Percentages
bars = ax2.bar(blood_types, number_of_donors, color=colors, edgecolor='black', linewidth=1.5)

# Add value labels and percentages on bars
for i, (bar, percentage) in enumerate(zip(bars, percentages)):
    height = bar.get_height()
    # Number of donors
    ax2.text(bar.get_x() + bar.get_width()/2., height + 0.1,
             f'{int(height)}', ha='center', va='bottom', fontsize=12, fontweight='bold')
    # Percentage
    ax2.text(bar.get_x() + bar.get_width()/2., height/2,
             f'{percentage:.1f}%', ha='center', va='center', fontsize=11, 
             color='white', fontweight='bold')

ax2.set_xlabel('Blood Type', fontsize=12, fontweight='bold')
ax2.set_ylabel('Number of Donors', fontsize=12, fontweight='bold')
ax2.set_title('Blood Type Donations (Bar Chart with %)', fontsize=14, fontweight='bold')
ax2.set_ylim(0, max(number_of_donors) + 2)
ax2.grid(True, alpha=0.3, axis='y')

# Add a text box with total donors
ax2.text(0.5, -0.15, f'Total Donors: {total_donors}', 
         transform=ax2.transAxes, ha='center', fontsize=12, 
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.show()

# Print detailed analysis
print("=" * 60)
print("BLOOD DONATION ANALYSIS")
print("=" * 60)

print(f"\nSUMMARY:")
print("-" * 40)
print(f"Total donors: {total_donors}")
print(f"\nBlood Type Distribution:")
print("-" * 40)
print(f"{'Blood Type':<12} {'Donors':<10} {'Percentage':<12}")
print("-" * 40)

for blood_type, count, percentage in zip(blood_types, number_of_donors, percentages):
    print(f"{blood_type:<12} {count:<10} {percentage:.1f}%")

print("-" * 40)

# Find most and least common
most_common_idx = np.argmax(number_of_donors)
least_common_idx = np.argmin(number_of_donors)

print(f"\nINSIGHTS:")
print("-" * 40)
print(f"• Most common blood type: Type {blood_types[most_common_idx]} "
      f"({number_of_donors[most_common_idx]} donors, {percentages[most_common_idx]:.1f}%)")
print(f"• Least common blood type: Type {blood_types[least_common_idx]} "
      f"({number_of_donors[least_common_idx]} donors, {percentages[least_common_idx]:.1f}%)")

# Compare with general population distribution (approximate)
general_population = {'A': 30, 'B': 20, 'O': 45, 'AB': 5}  # Approximate percentages

print(f"\nCOMPARISON WITH GENERAL POPULATION:")
print("-" * 40)
print(f"{'Blood Type':<12} {'Sample %':<10} {'Population %':<15} {'Difference':<10}")
print("-" * 40)

for blood_type, sample_pct in zip(blood_types, percentages):
    pop_pct = general_population[blood_type]
    diff = sample_pct - pop_pct
    print(f"{blood_type:<12} {sample_pct:.1f}%{'':<7} {pop_pct}%{'':<12} {diff:+.1f}%")

print("-" * 40)

# Conclusion
print(f"\nCONCLUSION:")
print("-" * 40)
print("• Type A blood was the most common donation, followed by Type O.")
print("• Type AB blood was the least common, which matches general population trends.")
print("• The clinic collected a good mix of all blood types.")
print(f"• Type O blood, which is often in high demand as the universal donor, ")
print(f"  made up {percentages[2]:.1f}% of donations (7 units).")
````

Use a list comprehension to turn the given list into np.arrays.

````python
import numpy as np
import matplotlib.pyplot as plt

# Data: [price, number_of_tickets]
prices_list = [[1, 1], [5, 10], [10, 15], [15, 20], [20, 25], [25, 30], [30, 35], [35, 40]]

# Convert to numpy arrays for easier manipulation
prices_list = np.array(prices_list)
ticket_prices = prices_list[:, 0]  # First column: price in dollars
ticket_quantities = prices_list[:, 1]  # Second column: number of tickets

# Create the scatter plot
plt.figure(figsize=(10, 6))

# Scatter plot
plt.scatter(ticket_prices, ticket_quantities, color='darkred', s=150, 
            alpha=0.7, edgecolor='black', linewidth=1.5, zorder=5)

# Add labels to each point
for i, (price, quantity) in enumerate(prices_list):
    plt.annotate(f'${price}', (price - 1.5, quantity + 0.5), 
                fontsize=10, fontweight='bold', color='darkred')
    plt.annotate(f'{quantity} tickets', (price - 1.5, quantity - 1.2), 
                fontsize=9, color='gray')

# Add a trend line to show the relationship
z = np.polyfit(ticket_prices, ticket_quantities, 1)
p = np.poly1d(z)
plt.plot(ticket_prices, p(ticket_prices), "b--", linewidth=1.5, 
         alpha=0.5, label=f'Trend line (slope = {z[0]:.2f})')

# Customize the plot
plt.xlabel('Ticket Price ($)', fontsize=12, fontweight='bold')
plt.ylabel('Number of Tickets', fontsize=12, fontweight='bold')
plt.title('Raffle Ticket Prices vs. Number of Tickets', fontsize=14, fontweight='bold')

# Set axis ticks
plt.xticks(ticket_prices, [f'${p}' for p in ticket_prices])
plt.yticks(ticket_quantities)

# Add grid
plt.grid(True, alpha=0.3, linestyle='--')

# Add legend
plt.legend(fontsize=10)

# Set axis limits for better view
plt.xlim(0, 40)
plt.ylim(0, 45)

plt.tight_layout()
plt.show()

# Print analysis
print("=" * 60)
print("RAFFLE TICKET PRICE ANALYSIS")
print("=" * 60)

print("\nTICKET OPTIONS:")
print("-" * 50)
print(f"{'Price ($)':<12} {'Tickets':<12} {'Cost per Ticket ($)':<20}")
print("-" * 50)

for price, quantity in prices_list:
    cost_per_ticket = price / quantity
    print(f"${price:<11} {quantity:<12} ${cost_per_ticket:.3f}")

print("-" * 50)

# Calculate statistics
print("\nSTATISTICS:")
print("-" * 50)
print(f"Lowest price: ${min(ticket_prices)} ({min(ticket_quantities)} tickets)")
print(f"Highest price: ${max(ticket_prices)} ({max(ticket_quantities)} tickets)")
print(f"Average cost per ticket: ${np.mean(ticket_prices / ticket_quantities):.3f}")

# Calculate correlation
correlation = np.corrcoef(ticket_prices, ticket_quantities)[0, 1]
print(f"Correlation coefficient: {correlation:.4f}")

# Observations
print("\nOBSERVATIONS:")
print("-" * 50)
print("• As ticket price increases, the number of tickets increases linearly")
print(f"• Perfect positive correlation ({correlation:.4f})")
print("• The cheapest option is $1 for 1 ticket ($1.00 per ticket)")
print("• The most economical option is $35 for 40 tickets ($0.875 per ticket)")
print("• Buying in bulk offers better value:")

best_value_idx = np.argmin(ticket_prices / ticket_quantities)
worst_value_idx = np.argmax(ticket_prices / ticket_quantities)
print(f"  - Best value: ${ticket_prices[best_value_idx]} for {ticket_quantities[best_value_idx]} tickets")
print(f"  - Worst value: ${ticket_prices[worst_value_idx]} for {ticket_quantities[worst_value_idx]} tickets")

# Calculate savings
best_cost_per_ticket = ticket_prices[best_value_idx] / ticket_quantities[best_value_idx]
worst_cost_per_ticket = ticket_prices[worst_value_idx] / ticket_quantities[worst_value_idx]
savings_per_ticket = worst_cost_per_ticket - best_cost_per_ticket
print(f"• You save ${savings_per_ticket:.3f} per ticket by buying the bulk option")

print("\nCONCLUSION:")
print("-" * 50)
print("The scatter plot shows a perfect linear relationship between")
print("ticket price and quantity. This is a typical volume discount")
print("strategy where buying more tickets gives you a better per-ticket price.")
````

Now make a graph of the price per ticket versus number of tickets bought.

````python
import numpy as np
import matplotlib.pyplot as plt

# Data: [price, number_of_tickets]
prices_list = [[1, 1], [5, 10], [10, 15], [15, 20], [20, 25], [25, 30], [30, 35], [35, 40]]

# Convert to numpy arrays
prices_list = np.array(prices_list)
ticket_prices = prices_list[:, 0]  # Total price in dollars
ticket_quantities = prices_list[:, 1]  # Number of tickets

# Calculate price per ticket
price_per_ticket = ticket_prices / ticket_quantities

# Create the plot
plt.figure(figsize=(10, 6))

# Scatter plot
plt.scatter(ticket_quantities, price_per_ticket, color='darkgreen', s=150, 
            alpha=0.7, edgecolor='black', linewidth=1.5, zorder=5)

# Add labels to each point
for i, (quantity, price) in enumerate(zip(ticket_quantities, price_per_ticket)):
    plt.annotate(f'${ticket_prices[i]} for {quantity} tickets', 
                (quantity + 0.5, price - 0.02), 
                fontsize=9, fontweight='bold', color='darkgreen')
    plt.annotate(f'${price:.3f}/ticket', 
                (quantity + 0.5, price + 0.01), 
                fontsize=8, color='gray')

# Add a trend line
z = np.polyfit(ticket_quantities, price_per_ticket, 2)  # Quadratic fit for curved trend
p = np.poly1d(z)
x_smooth = np.linspace(min(ticket_quantities), max(ticket_quantities), 100)
plt.plot(x_smooth, p(x_smooth), "r--", linewidth=1.5, alpha=0.5, 
         label='Trend line (decreasing cost)')

# Customize the plot
plt.xlabel('Number of Tickets Bought', fontsize=12, fontweight='bold')
plt.ylabel('Price per Ticket ($)', fontsize=12, fontweight='bold')
plt.title('Raffle Ticket: Price per Ticket vs. Quantity Bought', fontsize=14, fontweight='bold')

# Set x-axis ticks
plt.xticks(ticket_quantities)

# Add grid
plt.grid(True, alpha=0.3, linestyle='--')

# Add legend
plt.legend(fontsize=10)

# Set axis limits
plt.xlim(0, 45)
plt.ylim(0.8, 1.1)

plt.tight_layout()
plt.show()

# Print analysis
print("=" * 60)
print("PRICE PER TICKET ANALYSIS")
print("=" * 60)

print("\nPRICE PER TICKET BREAKDOWN:")
print("-" * 50)
print(f"{'Tickets':<12} {'Total Price':<15} {'Price/Ticket':<15} {'Savings vs $1 ticket':<20}")
print("-" * 50)

base_price = 1.0  # $1 for 1 ticket

for quantity, total_price, ppt in zip(ticket_quantities, ticket_prices, price_per_ticket):
    savings = (base_price - ppt) * quantity
    savings_percent = ((base_price - ppt) / base_price) * 100
    print(f"{quantity:<12} ${total_price:<14} ${ppt:.3f}{'':<11} {savings_percent:.1f}% (${savings:.2f} total)")

print("-" * 50)

# Calculate statistics
print("\nKEY INSIGHTS:")
print("-" * 50)

# Best and worst value
best_value_idx = np.argmin(price_per_ticket)
worst_value_idx = np.argmax(price_per_ticket)

print(f"• Best value: {ticket_quantities[best_value_idx]} tickets for ${ticket_prices[best_value_idx]}")
print(f"  → ${price_per_ticket[best_value_idx]:.3f} per ticket")
print(f"• Worst value: {ticket_quantities[worst_value_idx]} tickets for ${ticket_prices[worst_value_idx]}")
print(f"  → ${price_per_ticket[worst_value_idx]:.3f} per ticket")

# Savings calculations
max_savings = (base_price - min(price_per_ticket)) * max(ticket_quantities)
print(f"\nSAVINGS WITH BULK PURCHASE:")
print("-" * 50)
print(f"• Buying 1 ticket at a time: $1.00 per ticket")
print(f"• Buying {int(max(ticket_quantities))} tickets: ${min(price_per_ticket):.3f} per ticket")
print(f"• You save: ${base_price - min(price_per_ticket):.3f} per ticket")
print(f"• Total savings on {int(max(ticket_quantities))} tickets: ${max_savings:.2f}")

# Percentage savings
best_savings_percent = ((base_price - min(price_per_ticket)) / base_price) * 100
print(f"• That's a {best_savings_percent:.1f}% discount compared to buying single tickets")

print("\nCONCLUSION:")
print("-" * 50)
print("This graph clearly shows the volume discount strategy:")
print("• The price per ticket decreases as you buy more tickets")
print("• The relationship is non-linear - the biggest drop happens early")
print("• Best value is at the highest quantity (40 tickets for $35)")
print("• This encourages customers to spend more by offering better per-unit pricing")
````
