## R-numbers

Write markdown to recognize valid studentnumbers (R-numbers). A valid number starts with an R (case doesn't matter) and is followed by 7 numbers of which the first is 0 or 1.

Apply it to a list of strings and as a result print two lists: the valid ones and the invalid ones.

```python
the_list = ['r0003689', 'u0103828', 'R0001555', 'r001782', '0008750']
```

```python
import re

pattern = r"^[Rr][01][0-9]{6}$"

the_list = ['r0003689', 'u0103828', 'R0001555', 'r001782', '0008750']

valid = []
invalid = []

for item in the_list:
    if re.match(pattern, item):
        valid.append(item)
    else:
        invalid.append(item)

print("Valid:", valid)
print("Invalid:", invalid)
```

## Find text

Write a Python program to search some literals strings in a string.

Example:

> Sample text : 'The quick brown fox jumps over the lazy dog.'
> Searched words : 'fox', 'dog', 'horse'

Will render:

> Searching for "fox" in "The quick brown fox jumps over the lazy dog." -> Matched!                                                                                                      
> Searching for "dog" in "The quick brown fox jumps over the lazy dog." -> Matched!                                                                                                      
> Searching for "horse" in "The quick brown fox jumps over the lazy dog." -> Not Matched!

```python
import re

text = "The quick brown fox jumps over the lazy dog."
words = ['fox', 'dog', 'horse']

for word in words:
    if re.search(word, text):
        print(f'Searching for "{word}" in "{text}" -> Matched!')
    else:
        print(f'Searching for "{word}" in "{text}" -> Not Matched!')
```

## Casing

When using pascal casing, every word in a variable name starts with an uppercase letter. In case of snake casing, indivual words start with a lowercase letter and are separated by an underscore.  

Write a function to convert pascal case in snake case.

Example:

> "PythonExercises"

Will render:

> "python_exercises"

```python
import re

def pascal_to_snake(text):
    
    pattern = r'(?<!^)(?=[A-Z])'
    return re.sub(pattern, '_', text).lower()

# Test the function
test = "PythonExercises"
result = pascal_to_snake(test)
print(f"Original: {test}")
print(f"Snake case: {result}")
```

## Casing v2

Now do the same, but have the function detect what type is used (snake or pascal) and return the other.

Examples:

> "PythonExercises" -> "python_exercises"  
> "python_exercises" -> "PythonExercises"

```python
import re

def convert_case(text):
    # Detect if it's snake_case (contains underscore)
    if '_' in text:
        # Convert snake_case to PascalCase
        words = text.split('_')
        return ''.join(word.capitalize() for word in words)
    else:
        # Convert PascalCase to snake_case
        pattern = r'(?<!^)(?=[A-Z])'
        return re.sub(pattern, '_', text).lower()

# Test the function
test_cases = ["PythonExercises", "python_exercises"]

for test in test_cases:
    result = convert_case(test)
    print(f"Original: {test}")
    print(f"Converted: {result}")
    print("-" * 30)
```

## Whitespaces

Write a Python program to remove all whitespaces from a string.

```python
import re

def remove_whitespaces(text):
    
    return re.sub(r'\s+', '', text)

# Test the function
test_string = "  Hello   world!  \n  This has \t many  spaces.  "

result = remove_whitespaces(test_string)

print(f"Original: '{test_string}'")
print(f"Without whitespaces: '{result}'")
```

## Split

Write a Python program to split a string with multiple delimiters.

Note : A delimiter is a sequence of one or more characters used to specify the boundary between separate, independent regions in plain text or other data streams. An example of a delimiter is the comma character, which acts as a field delimiter in a sequence of comma-separated values.

Example:

> 'The quick brown\nfox jumps*over the lazy dog.' -> ['The quick brown', 'fox jumps', 'over the lazy dog.'] 

```python
import re

text = 'The quick brown\nfox jumps*over the lazy dog.'


pattern = r'\n|\*'

# Split the string
result = re.split(pattern, text)

print(f"Original: {text}")
print(f"Split result: {result}")
```

## Check numbers

Write a Python program to check if an input string is a number with maximally 2 digits behind the comma.

Examples:

> 123.11 -> True  
> 123.1 -> True  
> 123 -> True  
> 0.21 -> True  
> 123.1214 -> False  
> 3.124587 -> False  
> e666.86 -> False


```python
import re

def check_number(s):
    # Pattern: optional digits, optional decimal with 1-2 digits
    pattern = r'^\d+(\.\d{1,2})?$'
    return bool(re.match(pattern, s))

# Test cases
test_cases = ['123.11', '123.1', '123', '0.21', '123.1214', '3.124587', 'e666.86']

for test in test_cases:
    result = check_number(test)
    print(f"{test:10} -> {result}")
```

For this course, we give you the assert statements to test your code for free. So you just need to code a function is_nice() that checks whether the input string meets the criteria to be defined as nice.  

A complete input file for this challenge is available in the map 'Files'.

```python
import re

def is_nice(s):
    vowels = len(re.findall(r'[aeiou]', s)) >= 3
    double = re.search(r'(.)\1', s) is not None
    forbidden = not re.search(r'ab|cd|pq|xy', s)
    return vowels and double and forbidden

test_strings = [
    "ugknbfddgicrmopn",
    "aaa",
    "jchzalrnumimnmhp",
    "haegwjzuvuyypxyu",
    "dvszwmarrgswjxmb"
]

nice_count = 0
for s in test_strings:
    if is_nice(s):
        nice_count += 1
        print(f"{s} -> NICE")
    else:
        print(f"{s} -> NAUGHTY")

print(f"Number of nice strings: {nice_count}")
```

```python
# Test statements:
assert is_nice("ugknbfddgicrmopn") == True
assert is_nice("aaa") == True
assert is_nice("jchzalrnumimnmhp") == False
assert is_nice("haegwjzuvuyypxyu") == False
assert is_nice("dvszwmarrgswjxmb") == False
```

## Advent of code 2015, day 6

We won't be doing day 6 entirely, but the challenge starts with a fun little regex-problem.

The instructions can be found here: [advent of code 2015, day 6](https://adventofcode.com/2015/day/6)

The challenge starts with instructions formulated as follows:

* turn on 606,361 through 892,600
* turn off 448,208 through 645,684
* toggle 50,472 through 452,788

Your task is to refrase these lines into a datastructure that a function can handle.
So for this exercise you write a function translate_line() that turns the incoming string into a tuple. You can find the exact lay-out of the elements in the tuple in the given assert-statements to test your function.

A complete input file for this challenge is available in the map 'Files'.

```python
import re

def translate_line(line):
    # Pattern to match instruction and coordinates
    pattern = r'(turn on|turn off|toggle) (\d+),(\d+) through (\d+),(\d+)'
    match = re.match(pattern, line)
    
    if match:
        instruction = match.group(1)
        x1 = int(match.group(2))
        y1 = int(match.group(3))
        x2 = int(match.group(4))
        y2 = int(match.group(5))
        
        # Ensure coordinates are in order (x1 <= x2, y1 <= y2)
        x1, x2 = min(x1, x2), max(x1, x2)
        y1, y2 = min(y1, y2), max(y1, y2)
        
        return (instruction, x1, y1, x2, y2)
    return None

# Test the function
test_lines = [
    "turn on 606,361 through 892,600",
    "turn off 448,208 through 645,684",
    "toggle 50,472 through 452,788"
]

for line in test_lines:
    result = translate_line(line)
    print(f"{line} -> {result}")
```

```python
assert translate_line('turn on 606,361 through 892,600') == ('turn on', (606, 361), (892, 600))
assert translate_line('turn off 448,208 through 645,684') == ('turn off', (448, 208), (645, 684))
assert translate_line('toggle 50,472 through 452,788') == ('toggle', (50, 472), (452, 788))

# translate_line('Your code should generate an error now, one you threw yourself using "raise Exception".')
```
