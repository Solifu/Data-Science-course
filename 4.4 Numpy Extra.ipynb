# Exercises numpy

Numpy is embedded quite well into pandas (which is the next chapter). NumPy is very useful for performing logical and mathematical calculations on arrays and matrices. This tool performs these operations much faster and more efficiently than Python lists. 

Furthermore, Numpy uses less memory and storage space, which is the main advantage. This is because numpy is typed. Normal Python is typeless, so if you create a list of integers, Python ensures that it is also suitable for floating point numbers or very long strings. The more items you add to the list, the more space is used as a list grows exponentially. This means that a lot of memory is required, whereas numpy knows what the array is intended for and allocates exactly enough memory.

## Rotate an array

Create a python list with 3 items, and initiate every item with a list of 4 items. As content of the list, use the next letter of the alphabet. ([tip](https://www.askpython.com/python/built-in-methods/python-chr-and-ord-methods)).

To generate:
```
[['a', 'b', 'c', 'd'], ['e', 'f', 'g', 'h'], ['i', 'j', 'k', 'l']]
```

````python

# Method 1: Using chr() in a loop
original_list = []
letter_code = 97  # 'a' starts at 97

for row in range(3):
    row_list = []
    for col in range(4):
        row_list.append(chr(letter_code))
        letter_code += 1
    original_list.append(row_list)

print("Original list (3x4):")
print(original_list)

# Rotate the list using zip()

rotated_list = [list(row) for row in zip(*original_list)]

print("Original list:")
print(original_list)
print("\nRotated list (4x3):")
print(rotated_list)
````

Now turn the array so you get a rotated list of lists. This would be a list of 4 items, each of them containing a list with 3 items. (Do note however that if we were to change to original list to a 5x12-matrix your code should still work.)

Output:
```
[['a', 'e', 'i'], ['b', 'f', 'j'], ['c', 'g', 'k'], ['d', 'h', 'l']]
```

````python
import numpy as np

matrix = [['a', 'b', 'c', 'd'], 
          ['e', 'f', 'g', 'h'], 
          ['i', 'j', 'k', 'l']]

transposed = np.array(matrix).T.tolist()
print(transposed)
````

````python
import numpy as np
import re

# Use the sample data directly instead of reading from a file
data = """
turn on 0,0 through 999,999
toggle 0,0 through 999,0
turn off 499,499 through 500,500
"""

# Create a 1000x1000 grid of zeros (lights off)
grid = np.zeros((1000, 1000), dtype=bool)

# Process each line
for line in data.strip().split('\n'):
    line = line.strip()
    if not line:
        continue
    
    # Regex to parse instruction
    match = re.match(r'(turn on|turn off|toggle) (\d+),(\d+) through (\d+),(\d+)', line)
    if match:
        command = match.group(1)
        x1, y1, x2, y2 = map(int, match.groups()[1:])
        
        # Ensure coordinates are in order (smaller to larger)
        x1, x2 = sorted([x1, x2])
        y1, y2 = sorted([y1, y2])
        
        # Apply the command to the grid slice
        if command == 'turn on':
            grid[x1:x2+1, y1:y2+1] = True
        elif command == 'turn off':
            grid[x1:x2+1, y1:y2+1] = False
        elif command == 'toggle':
            grid[x1:x2+1, y1:y2+1] = ~grid[x1:x2+1, y1:y2+1]

# Count how many lights are lit
lit_count = np.sum(grid)
print(f"Number of lights lit: {lit_count}")
````

* the action
* tuple of start coordinate
* tuple of end coordinate

It will apply the action to all points within the coordinates.

````python
import numpy as np
import re

def get_value(current, action):
   
    if action == 'turn on':
        return 1
    elif action == 'turn off':
        return 0
    elif action == 'toggle':
        return 1 - current
    else:
        raise ValueError(f"Unknown action: {action}")

def apply_to_grid(line, grid):
    
    action, start_coord, end_coord = line
    
    # Unpack coordinates
    x1, y1 = start_coord
    x2, y2 = end_coord
    
    # Ensure coordinates are in order
    x1, x2 = sorted([x1, x2])
    y1, y2 = sorted([y1, y2])
    
    # Apply the action to each light in the rectangle
    for x in range(x1, x2 + 1):
        for y in range(y1, y2 + 1):
            grid[x, y] = get_value(grid[x, y], action)
    
    return grid

def parse_instruction(line):
    
    match = re.match(r'(turn on|turn off|toggle) (\d+),(\d+) through (\d+),(\d+)', line)
    if match:
        action = match.group(1)
        x1, y1 = int(match.group(2)), int(match.group(3))
        x2, y2 = int(match.group(4)), int(match.group(5))
        return [action, (x1, y1), (x2, y2)]
    return None

# Create the 1000x1000 grid (all lights off initially)
grid = np.zeros((1000, 1000), dtype=int)

# Your data list (replace this with your actual instructions)
# For demonstration, using the sample instructions:
data = """
turn on 0,0 through 999,999
toggle 0,0 through 999,0
turn off 499,499 through 500,500
"""

# Parse all instructions into a list
instructions = []
for line in data.strip().split('\n'):
    if line.strip():
        instruction = parse_instruction(line.strip())
        if instruction:
            instructions.append(instruction)

# Apply all instructions to the grid
print("Processing instructions...")
for i, instruction in enumerate(instructions):
    print(f"Processing instruction {i+1}/{len(instructions)}: {instruction[0]}")
    grid = apply_to_grid(instruction, grid)

# Calculate and print the sum (number of lights turned on)
lights_on = np.sum(grid)
print(f"\n{'='*50}")
print(f"Number of lights turned on: {lights_on}")
print(f"{'='*50}")

# Optional: Show some statistics
print(f"\nGrid statistics:")
print(f"  Shape: {grid.shape}")
print(f"  Total lights: {grid.size}")
print(f"  Lights on: {lights_on}")
print(f"  Lights off: {grid.size - lights_on}")
print(f"  Percentage on: {100 * lights_on / grid.size:.2f}%")
````
