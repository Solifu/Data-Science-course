Cell 1: Create an original list

```python

# Method 1: Using chr() in a loop
original_list = []
letter_code = 97  # 'a' starts at 97

for row in range(3):
    row_list = []
    for col in range(4):
        row_list.append(chr(letter_code))
        letter_code += 1
    original_list.append(row_list)

print("Original list (3x4):")
print(original_list)
```

Cell 2: Rotate

```python
# Rotate the list using zip()

rotated_list = [list(row) for row in zip(*original_list)]

print("Original list:")
print(original_list)
print("\nRotated list (4x3):")
print(rotated_list)
```

Cell 3: Version using numpy

```python
# Using numpy for rotation

import numpy as np

# Convert to numpy array
np_array = np.array(original_list)

print("Original numpy array:")
print(np_array)

# Rotate 90 degrees
rotated_np = np.rot90(np_array)

print("\nRotated 90° (numpy):")
print(rotated_np)
```

Exercise 2

Cell 1: Create grid and helper function

```python
import numpy as np

# Create 1000x1000 grid (0=off, 1=on)
lights = np.zeros((1000, 1000), dtype=int)

def new_value(current, action):
    if action == 'turn on': return 1
    if action == 'turn off': return 0
    if action == 'toggle': return 1 - current

print("Grid ready!")
```

Cell 2 - Simple apply function:

```python
def apply(action, x1, y1, x2, y2):
    for x in range(x1, x2+1):
        for y in range(y1, y2+1):
            lights[x][y] = new_value(lights[x][y], action)
    
    print(f"{action}: ({x1},{y1}) to ({x2},{y2}) - Lights on: {np.sum(lights)}")
```

Cell 3 - Test with examples:

```python
# Reset
lights = np.zeros((1000, 1000), dtype=int)

# Test instructions
print("Testing examples:\n")
apply('turn on', 0, 0, 999, 999)
apply('toggle', 0, 0, 999, 0)
apply('turn off', 499, 499, 500, 500)

print(f"\nFINAL: {np.sum(lights)} lights on")
```

Cell 4 - Load real instructions:

```python
import re

# Read file
with open('input.txt', 'r') as f:
    lines = f.readlines()

# Reset lights
lights = np.zeros((1000, 1000), dtype=int)

# Process each line
for line in lines:
    line = line.strip()
    if 'turn on' in line:
        nums = re.findall(r'\d+', line)
        apply('turn on', int(nums[0]), int(nums[1]), int(nums[2]), int(nums[3]))
    elif 'turn off' in line:
        nums = re.findall(r'\d+', line)
        apply('turn off', int(nums[0]), int(nums[1]), int(nums[2]), int(nums[3]))
    elif 'toggle' in line:
        nums = re.findall(r'\d+', line)
        apply('toggle', int(nums[0]), int(nums[1]), int(nums[2]), int(nums[3]))

print(f"\nFINAL ANSWER: {np.sum(lights)} lights on")
```
