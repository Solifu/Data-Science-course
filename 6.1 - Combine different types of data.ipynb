Step 1: convert the aircraftType-column to a Series. Concat these two series to the original dataframe from which you can immediately drop the aircraftType-column.

````python
# Load the pickle files
all_flights = pickle.load(open("files/all_flights.p", "rb"))
all_aircraftTypes = pickle.load(open("files/all_aircraftTypes.p", "rb"))
all_airlines = pickle.load(open("files/all_airlines.p", "rb"))
all_destinations = pickle.load(open("files/all_destinations.p", "rb"))

# Examine the structure of aircraftType column
print("Original all_flights columns:", list(all_flights.columns))
print("\nFirst row aircraftType:")
print(all_flights['aircraftType'].iloc[0])
print("\nType:", type(all_flights['aircraftType'].iloc[0]))

# Convert the aircraftType column to a Series and expand it
# If it's a dictionary column
if isinstance(all_flights['aircraftType'].iloc[0], dict):
    aircraft_expanded = all_flights['aircraftType'].apply(pd.Series)
else:
    # If it's a string representation of a dictionary
    import ast
    aircraft_expanded = all_flights['aircraftType'].apply(
        lambda x: pd.Series(ast.literal_eval(x) if isinstance(x, str) else x)
    )

print("\nExpanded aircraft columns:", list(aircraft_expanded.columns))

# Concat the expanded series to the original dataframe (dropping the original column)
all_flights_clean = pd.concat([
    all_flights.drop('aircraftType', axis=1),
    aircraft_expanded
], axis=1)

print("\nNew all_flights_clean columns:", list(all_flights_clean.columns))
print("\nFirst 2 rows:")
print(all_flights_clean.head(2))

# Now you can merge with all_aircraftTypes if needed
# Assuming they share a common key like 'code' or 'icao'
if 'icao' in all_flights_clean.columns and 'icao' in all_aircraftTypes.columns:
    all_flights_final = pd.merge(
        all_flights_clean,
        all_aircraftTypes,
        on='icao',
        how='left'
    )
    print("\nAfter merge with all_aircraftTypes:")
    print(all_flights_final.head(2))
````

Step 2: merge the two new columns with the aircraft_types-dataframe to get the extended information on the plane.

````python
print("\n" + "=" * 60)
print("STEP 2: Merging with aircraft_types dataframe")
print("-" * 60)

# Identify common columns to merge on
print(f"all_flights_clean columns: {list(all_flights_clean.columns)}")
print(f"all_aircraftTypes columns: {list(all_aircraftTypes.columns)}")

# Find common keys - typically 'icao' or 'code'
possible_keys = ['icao', 'code', 'aircraftCode', 'type']
merge_key = None

for key in possible_keys:
    if key in all_flights_clean.columns and key in all_aircraftTypes.columns:
        merge_key = key
        break

if merge_key:
    print(f"\nMerging on: '{merge_key}'")
    
    # Perform the merge
    all_flights_final = pd.merge(
        all_flights_clean,
        all_aircraftTypes,
        on=merge_key,
        how='left'
    )
    
    print(f"\nFinal merged shape: {all_flights_final.shape}")
    print(f"Final columns: {list(all_flights_final.columns)}")
    print("\nFirst 2 rows of final dataframe:")
    print(all_flights_final.head(2))
    
    # Check if merge was successful
    print(f"\nRows before merge: {len(all_flights_clean)}")
    print(f"Rows after merge: {len(all_flights_final)}")
    
    # Count how many rows got matched
    matched = all_flights_final[all_flights_final['manufacturer'].notna() if 'manufacturer' in all_flights_final.columns else True].shape[0]
    print(f"Rows matched with aircraft data: {matched}")
    
else:
    print("\nNo common merge key found. Trying alternative approach...")
    
    # Alternative: If aircraft_expanded contains the key
    if 'icao' in aircraft_expanded.columns and 'icao' in all_aircraftTypes.columns:
        all_flights_final = pd.merge(
            all_flights_clean,
            all_aircraftTypes,
            left_on='icao',
            right_on='icao',
            how='left'
        )
        print(f"Merged using 'icao' from expanded columns")
        print(f"Final shape: {all_flights_final.shape}")
    else:
        print("Could not determine merge key. Please check column names.")
````

The airlines are an easy one. Simply merge based on "prefixICAO" in flights. This is "icao" in all_airlines. 

````python
# Load the pickle files
all_flights = pickle.load(open("files/all_flights.p", "rb"))
all_airlines = pickle.load(open("files/all_airlines.p", "rb"))

print("=" * 60)
print("MERGING AIRLINES DATA WITH FLIGHTS")
print("=" * 60)

# Check the data
print("all_flights columns:", list(all_flights.columns))
print("\nall_airlines columns:", list(all_airlines.columns))

# Verify the merge keys
print(f"\nMerge keys: all_flights['prefixICAO'] -> all_airlines['icao']")

# Perform the merge
flights_with_airlines = pd.merge(
    all_flights,
    all_airlines,
    left_on='prefixICAO',
    right_on='icao',
    how='left'
)

print(f"\nOriginal flights shape: {all_flights.shape}")
print(f"New merged shape: {flights_with_airlines.shape}")

# Show results
print("\nFirst 2 rows after merge:")
print(flights_with_airlines.head(2))

# Check merge quality
matched = flights_with_airlines['name'].notna().sum() if 'name' in flights_with_airlines.columns else 0
print(f"\nRows with airline name matched: {matched} out of {len(flights_with_airlines)}")

# If you want to drop the redundant 'icao' column from airlines
if 'icao' in flights_with_airlines.columns and 'prefixICAO' in flights_with_airlines.columns:
    flights_with_airlines = flights_with_airlines.drop('icao', axis=1)

print("\nFinal columns:", list(flights_with_airlines.columns))
````
