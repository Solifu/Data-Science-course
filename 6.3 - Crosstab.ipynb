# Cross tabulation

Put the days in chronological order!

```python


# Import data
df = pd.read_csv("files/tips.csv")
print("Original data - first 5 rows:")
print(df.head())

# Check unique days
print("\nUnique days in original order:")
print(df['day'].unique())

# Define chronological order (Thursday to Sunday)
chronological_order = ['Thur', 'Fri', 'Sat', 'Sun']

# Convert day to categorical with specified order
df['day'] = pd.Categorical(df['day'], categories=chronological_order, ordered=True)

# Sort by day
df_sorted = df.sort_values('day')

print("\n" + "="*70)
print("DATA WITH DAYS IN CHRONOLOGICAL ORDER:")
print("="*70)
print(df_sorted.head(10))

# Verify the order
print("\nUnique days after sorting:")
print(df_sorted['day'].unique())

# Optional: Sort by day and then by time
df_sorted = df.sort_values(['day', 'time'])
print("\n" + "="*70)
print("SORTED BY DAY THEN TIME:")
print("="*70)
print(df_sorted[['day', 'time', 'total_bill', 'tip']].head(10))
```

List how many tips the waiter received per day during lunch or dinner.

```python


# Import data
df = pd.read_csv("files/tips.csv")

# Define chronological order for days
chronological_order = ['Thur', 'Fri', 'Sat', 'Sun']
df['day'] = pd.Categorical(df['day'], categories=chronological_order, ordered=True)

# Count tips per day during lunch or dinner
# Method 1: Using crosstab
tips_count = pd.crosstab(df['day'], df['time'])

print("TIPS COUNT PER DAY BY MEAL TIME")
print("=" * 50)
print(tips_count)
print()

# Method 2: Using groupby
tips_count_group = df.groupby(['day', 'time']).size().unstack()
print("Using groupby (same result):")
print(tips_count_group)
print()

# Method 3: Add totals
tips_count_with_totals = tips_count.copy()
tips_count_with_totals['Total'] = tips_count.sum(axis=1)
tips_count_with_totals.loc['Total'] = tips_count.sum()

print("WITH TOTALS:")
print("=" * 50)
print(tips_count_with_totals)
```

You can calculate the row and column totals by using the margins attribute.  

```python


# Import data
df = pd.read_csv("files/tips.csv")

# Define chronological order for days
chronological_order = ['Thur', 'Fri', 'Sat', 'Sun']
df['day'] = pd.Categorical(df['day'], categories=chronological_order, ordered=True)

# Count tips per day during lunch or dinner with margins
tips_count = pd.crosstab(
    df['day'], 
    df['time'], 
    margins=True, 
    margins_name='Total'
)

print("TIPS COUNT PER DAY BY MEAL TIME (WITH MARGINS)")
print("=" * 60)
print(tips_count)
```

Next, create a crosstab displaying the number of tips received per gender and per smoker type (rows) per time and per day (columns)

```python

# Import data
df = pd.read_csv("files/tips.csv")

# Define chronological order for days
chronological_order = ['Thur', 'Fri', 'Sat', 'Sun']
df['day'] = pd.Categorical(df['day'], categories=chronological_order, ordered=True)

# Create crosstab with multiple rows (sex, smoker) and columns (time, day)
tips_crosstab = pd.crosstab(
    index=[df['sex'], df['smoker']],
    columns=[df['time'], df['day']],
    margins=True,
    margins_name='Total'
)

print("TIPS COUNT PER GENDER AND SMOKER TYPE BY TIME AND DAY")
print("=" * 80)
print(tips_crosstab)
```

Besides getting counts (and percentages) by categories, we can also see aggregations of numeric variables by the categories. For instance, show the average tip amount by day and time.

```python


# Import data
df = pd.read_csv("files/tips.csv")

# Define chronological order for days
chronological_order = ['Thur', 'Fri', 'Sat', 'Sun']
df['day'] = pd.Categorical(df['day'], categories=chronological_order, ordered=True)

# Calculate average tip amount by day and time
# Method 1: Using pivot_table
avg_tip = pd.pivot_table(
    df,
    values='tip',
    index='day',
    columns='time',
    aggfunc='mean',
    margins=True,
    margins_name='Total'
)

print("AVERAGE TIP AMOUNT BY DAY AND TIME")
print("=" * 60)
print(avg_tip.round(2))
```

Display mean and median

```python

# Import data
df = pd.read_csv("files/tips.csv")

# Define chronological order for days
chronological_order = ['Thur', 'Fri', 'Sat', 'Sun']
df['day'] = pd.Categorical(df['day'], categories=chronological_order, ordered=True)

# Calculate mean and median tip amount by day and time
tip_stats = pd.pivot_table(
    df,
    values='tip',
    index='day',
    columns='time',
    aggfunc=['mean', 'median'],
    margins=True,
    margins_name='Total'
)

print("MEAN AND MEDIAN TIP AMOUNT BY DAY AND TIME")
print("=" * 70)
print(tip_stats.round(2))
```

Instead of using crosstab(), you can also use the function pivot_table().  So now use this function to show the number of tips received per time and per day.

```python


# Import data
df = pd.read_csv("files/tips.csv")

# Define chronological order for days
chronological_order = ['Thur', 'Fri', 'Sat', 'Sun']
df['day'] = pd.Categorical(df['day'], categories=chronological_order, ordered=True)

# Use pivot_table to count tips per time and per day
tip_count = pd.pivot_table(
    df,
    values='tip',           # Any column works for counting
    index='time',
    columns='day',
    aggfunc='count',        # Count number of occurrences
    margins=True,
    margins_name='Total',
    fill_value=0            # Fill NaN with 0
)

print("NUMBER OF TIPS PER TIME AND DAY")
print("=" * 60)
print(tip_count)
```
