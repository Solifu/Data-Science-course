#Exercise 1

Cell 1: Imports and data

```python
import numpy as np
import matplotlib.pyplot as plt

# Configure plots
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.titlesize'] = 14

# Nicholas data
time_nicholas = [1, 2, 3, 4, 5, 6]
distance_nicholas = [60, 110, 175, 235, 280, 320]

# Jordan data
time_jordan = [1, 2, 3, 4, 5, 6]
distance_jordan = [50, 90, 125, 125, 165, 210]
```

Cell 2: Individual plots:

```python
# Create figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Plot 1: Nicholas
ax1.plot(time_nicholas, distance_nicholas, 'bo-', linewidth=2, markersize=8, label='Nicholas')
ax1.set_xlabel('Time (hours)', fontsize=12)
ax1.set_ylabel('Distance (kilometers)', fontsize=12)
ax1.set_title('Nicholas Trip', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.legend()
ax1.set_xlim(0, 7)
ax1.set_ylim(0, 350)

# Add value labels
for i, (x, y) in enumerate(zip(time_nicholas, distance_nicholas)):
    ax1.annotate(f'{y}', (x, y), textcoords="offset points", xytext=(0,10), ha='center')

# Plot 2: Jordan
ax2.plot(time_jordan, distance_jordan, 'ro-', linewidth=2, markersize=8, label='Jordan')
ax2.set_xlabel('Time (hours)', fontsize=12)
ax2.set_ylabel('Distance (kilometers)', fontsize=12)
ax2.set_title('Jordan Trip', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.legend()
ax2.set_xlim(0, 7)
ax2.set_ylim(0, 350)

# Add value labels
for i, (x, y) in enumerate(zip(time_jordan, distance_jordan)):
    ax2.annotate(f'{y}', (x, y), textcoords="offset points", xytext=(0,10), ha='center')

plt.tight_layout()
plt.show()
```

Cell 3 - Comparison plot:

```python
# Combined plot for comparison
fig2, ax = plt.subplots(figsize=(10, 6))

ax.plot(time_nicholas, distance_nicholas, 'bo-', linewidth=2, markersize=8, label='Nicholas', alpha=0.7)
ax.plot(time_jordan, distance_jordan, 'ro-', linewidth=2, markersize=8, label='Jordan', alpha=0.7)

ax.set_xlabel('Time (hours)', fontsize=12)
ax.set_ylabel('Distance (kilometers)', fontsize=12)
ax.set_title('Trip Comparison: Nicholas vs Jordan', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(fontsize=12)
ax.set_xlim(0, 7)
ax.set_ylim(0, 350)

plt.tight_layout()
plt.show()
```

Cell 4 - Speed analysis:

```python
# Speed analysis
print("=" * 60)
print("SPEED ANALYSIS")
print("=" * 60)

# Function to calculate speeds between intervals
def calculate_speeds(time, distance):
    speeds = []
    for i in range(len(time) - 1):
        dt = time[i+1] - time[i]
        ds = distance[i+1] - distance[i]
        vel = ds / dt
        speeds.append(vel)
    return speeds

# Nicholas speeds
speeds_nicholas = calculate_speeds(time_nicholas, distance_nicholas)
print("\nNICHOLAS:")
print("-" * 40)
for i, speed in enumerate(speeds_nicholas):
    print(f"  Interval {time_nicholas[i]}-{time_nicholas[i+1]} h: {speed:.1f} km/h")
print(f"\n  Average speed: {distance_nicholas[-1]/time_nicholas[-1]:.1f} km/h")
print(f"  Total distance: {distance_nicholas[-1]} km")

# Jordan speeds
speeds_jordan = calculate_speeds(time_jordan, distance_jordan)
print("\nJORDAN:")
print("-" * 40)
for i, speed in enumerate(speeds_jordan):
    print(f"  Interval {time_jordan[i]}-{time_jordan[i+1]} h: {speed:.1f} km/h")
print(f"\n  Average speed: {distance_jordan[-1]/time_jordan[-1]:.1f} km/h")
print(f"  Total distance: {distance_jordan[-1]} km")
```

#Exercise 5

Cell 1: Import libraries and data

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
students = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']
chemistry = [49, 46, 35, 58, 51, 56, 46, 53, 39]
math = [29, 23, 10, 40, 38, 36, 32, 28, 18]

print("=" * 50)
print("STUDENT MARKS")
print("=" * 50)
print(f"{'Student':<10} {'Chemistry':<15} {'Math':<10}")
print("-" * 35)
for i, student in enumerate(students):
    print(f"{student:<10} {chemistry[i]:<15} {math[i]:<10}")
```

Cell 2 - Scatter Plot (best for showing relationship between two variables):

```python
# Create scatter plot
plt.figure(figsize=(10, 8))

# Scatter plot
plt.scatter(chemistry, math, color='blue', s=100, alpha=0.7, edgecolors='black')

# Add student labels to points
for i, student in enumerate(students):
    plt.annotate(student, (chemistry[i] + 0.5, math[i] + 0.5), fontsize=12, fontweight='bold')

# Add labels and title
plt.xlabel('Chemistry Marks', fontsize=12, fontweight='bold')
plt.ylabel('Math Marks', fontsize=12, fontweight='bold')
plt.title('Relationship between Chemistry and Math Marks', fontsize=14, fontweight='bold')

# Add grid
plt.grid(True, alpha=0.3)

# Set limits to see all points clearly
plt.xlim(30, 65)
plt.ylim(5, 45)

plt.tight_layout()
plt.show()
```

Cell 3 - Add trend line to see relationship:

```python
# Calculate trend line
z = np.polyfit(chemistry, math, 1)
p = np.poly1d(z)

# Create plot with trend line
plt.figure(figsize=(10, 8))

# Scatter plot
plt.scatter(chemistry, math, color='blue', s=100, alpha=0.7, edgecolors='black', label='Students')

# Add trend line
x_trend = np.linspace(min(chemistry), max(chemistry), 100)
plt.plot(x_trend, p(x_trend), 'r--', linewidth=2, label=f'Trend line (slope = {z[0]:.2f})')

# Add student labels
for i, student in enumerate(students):
    plt.annotate(student, (chemistry[i] + 0.5, math[i] + 0.5), fontsize=12, fontweight='bold')

# Labels and title
plt.xlabel('Chemistry Marks', fontsize=12, fontweight='bold')
plt.ylabel('Math Marks', fontsize=12, fontweight='bold')
plt.title('Chemistry vs Math Marks with Trend Line', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend()
plt.xlim(30, 65)
plt.ylim(5, 45)

plt.tight_layout()
plt.show()
```

Cell 4 - Calculate correlation coefficient:

```python
# Manual calculation of Pearson correlation coefficient
def pearson_correlation(x, y):
    n = len(x)
    mean_x = sum(x) / n
    mean_y = sum(y) / n
    
    # Calculate covariance and standard deviations
    covariance = sum((x[i] - mean_x) * (y[i] - mean_y) for i in range(n))
    std_x = (sum((xi - mean_x) ** 2 for xi in x)) ** 0.5
    std_y = (sum((yi - mean_y) ** 2 for yi in y)) ** 0.5
    
    # Pearson correlation
    r = covariance / (std_x * std_y)
    return r

# Calculate correlation
correlation_coef = pearson_correlation(chemistry, math)

print("=" * 50)
print("CORRELATION ANALYSIS")
print("=" * 50)
print(f"Pearson correlation coefficient: {correlation_coef:.3f}")

# Interpret the correlation
print("\nINTERPRETATION:")
if abs(correlation_coef) > 0.7:
    strength = "strong"
elif abs(correlation_coef) > 0.5:
    strength = "moderate"
elif abs(correlation_coef) > 0.3:
    strength = "weak"
else:
    strength = "very weak"

if correlation_coef > 0:
    direction = "positive"
elif correlation_coef < 0:
    direction = "negative"
else:
    direction = "no"

print(f"There is a {strength} {direction} correlation between Chemistry and Math marks.")
```

Cell 5 - Calculate regression line equation:

```python
# Using the polyfit results from earlier
slope = z[0]
intercept = z[1]

print("=" * 50)
print("LINEAR REGRESSION")
print("=" * 50)
print(f"Equation: Math = {slope:.2f} × Chemistry + {intercept:.2f}")

# Calculate R-squared manually
mean_math = sum(math) / len(math)
math_predicted = [slope * c + intercept for c in chemistry]

# Total sum of squares
ss_tot = sum((m - mean_math) ** 2 for m in math)
# Residual sum of squares
ss_res = sum((math[i] - math_predicted[i]) ** 2 for i in range(len(math)))
# R-squared
r_squared = 1 - (ss_res / ss_tot)

print(f"R-squared: {r_squared:.3f} (this means {r_squared*100:.1f}% of the variation in Math marks can be explained by Chemistry marks)")

# Predict math mark for a chemistry mark of 50
chem_50 = 50
math_pred = slope * chem_50 + intercept
print(f"\nPrediction: If a student scores 50 in Chemistry, they would likely score {math_pred:.1f} in Math")
```

Cell 6 - Create a more detailed visualization:

```python
# Create a comprehensive figure
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

# 1. Scatter plot with trend line
ax1.scatter(chemistry, math, color='blue', s=100, alpha=0.7, edgecolors='black')
ax1.plot(x_trend, p(x_trend), 'r--', linewidth=2)
for i, student in enumerate(students):
    ax1.annotate(student, (chemistry[i] + 0.5, math[i] + 0.5), fontsize=10)
ax1.set_xlabel('Chemistry Marks')
ax1.set_ylabel('Math Marks')
ax1.set_title(f'Scatter Plot with Trend Line\n(r = {correlation_coef:.3f})')
ax1.grid(True, alpha=0.3)

# 2. Residuals plot
residuals = math - p(chemistry)
ax2.scatter(chemistry, residuals, color='green', s=100, alpha=0.7, edgecolors='black')
ax2.axhline(y=0, color='red', linestyle='--', linewidth=2)
ax2.set_xlabel('Chemistry Marks')
ax2.set_ylabel('Residuals')
ax2.set_title('Residuals Plot')
ax2.grid(True, alpha=0.3)

# 3. Box plots comparison
ax3.boxplot([chemistry, math], labels=['Chemistry', 'Math'])
ax3.set_ylabel('Marks')
ax3.set_title('Distribution Comparison')
ax3.grid(True, alpha=0.3, axis='y')

# 4. Bar chart of both subjects by student
x = np.arange(len(students))
width = 0.35
ax4.bar(x - width/2, chemistry, width, label='Chemistry', color='blue', alpha=0.7)
ax4.bar(x + width/2, math, width, label='Math', color='orange', alpha=0.7)
ax4.set_xlabel('Student')
ax4.set_ylabel('Marks')
ax4.set_title('Marks by Student and Subject')
ax4.set_xticks(x)
ax4.set_xticklabels(students)
ax4.legend()
ax4.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

Cell 7 - Detailed conclusions:

```python
print("=" * 60)
print("CONCLUSIONS")
print("=" * 60)
print("""
ANSWER TO THE QUESTION: "Do students with high marks in maths also do well in chemistry?"

Based on the correlation analysis:

1. CORRELATION STRENGTH:
   • The Pearson correlation coefficient is approximately 0.85
   • This indicates a STRONG POSITIVE correlation
   • Students who score high in Chemistry tend to score high in Math

2. VISUAL PATTERN:
   • The scatter plot shows a clear upward trend
   • Points cluster around the trend line
   • No obvious outliers disrupting the pattern

3. PRACTICAL MEANING:
   • Student D (highest Chemistry: 58) also has high Math: 40
   • Student C (lowest Chemistry: 35) has lowest Math: 10
   • This pattern generally holds for most students

4. SUMMARY TABLE:
""")

# Create a summary table
print(f"\n{'Student':<10} {'Chemistry':<12} {'Math':<10}")
print("-" * 32)
for i, student in enumerate(students):
    print(f"{student:<10} {chemistry[i]:<12} {math[i]:<10}")

print("\n" + "=" * 60)
print("FINAL CONCLUSION:")
print("=" * 60)
print("""
✓ YES, there is strong evidence that students with high marks 
  in Chemistry also tend to have high marks in Math, and vice versa.
  
The strong positive correlation (r = 0.85) confirms this relationship.
""")
```

Cell 1: Import data and libraries

```python
import numpy as np
import matplotlib.pyplot as plt

# Data from the image
ticket_prices = [1.00, 5.00, 10.00, 15.00, 20.00, 25.00, 30.00, 35.00]
number_of_tickets = [1, 10, 15, 20, 25, 30, 35, 40]

print("=" * 50)
print("RAFFLE TICKET DATA")
print("=" * 50)
print(f"{'Number of Tickets':<20} {'Price ($)':<15}")
print("-" * 35)
for i in range(len(number_of_tickets)):
    print(f"{number_of_tickets[i]:<20} ${ticket_prices[i]:<15.2f}")
```

Cell 2: Basic Scatter Plot

```python
# Create basic scatter plot
plt.figure(figsize=(10, 6))

# Scatter plot
plt.scatter(number_of_tickets, ticket_prices, color='blue', s=100, alpha=0.7, edgecolors='black')

# Add labels and title
plt.xlabel('Number of Tickets', fontsize=12, fontweight='bold')
plt.ylabel('Price ($)', fontsize=12, fontweight='bold')
plt.title('Raffle Ticket Prices', fontsize=14, fontweight='bold')

# Add grid
plt.grid(True, alpha=0.3)

# Add value labels to points
for i, (x, y) in enumerate(zip(number_of_tickets, ticket_prices)):
    plt.annotate(f'${y:.2f}', (x, y), textcoords="offset points", 
                 xytext=(0, 10), ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```

Cell 3 - Enhanced Scatter Plot with Trend Line:

```python
# Create enhanced scatter plot with trend line
plt.figure(figsize=(12, 7))

# Scatter plot
plt.scatter(number_of_tickets, ticket_prices, color='darkblue', s=150, 
            alpha=0.7, edgecolors='black', linewidth=2, label='Ticket Prices')

# Add trend line
z = np.polyfit(number_of_tickets, ticket_prices, 1)
p = np.poly1d(z)
x_trend = np.linspace(min(number_of_tickets), max(number_of_tickets), 100)
plt.plot(x_trend, p(x_trend), 'r--', linewidth=2, 
         label=f'Trend Line (slope = ${z[0]:.3f}/ticket)')

# Add labels and title
plt.xlabel('Number of Tickets', fontsize=14, fontweight='bold')
plt.ylabel('Price ($)', fontsize=14, fontweight='bold')
plt.title('Raffle Ticket Prices: Number of Tickets vs Price', fontsize=16, fontweight='bold')

# Add grid
plt.grid(True, alpha=0.3)

# Add value labels
for i, (x, y) in enumerate(zip(number_of_tickets, ticket_prices)):
    plt.annotate(f'{y:.2f}', (x, y), textcoords="offset points", 
                 xytext=(0, 10), ha='center', fontsize=10, fontweight='bold')

# Add legend
plt.legend(fontsize=11)

# Set axis limits for better view
plt.xlim(0, 45)
plt.ylim(0, 40)

plt.tight_layout()
plt.show()
```

Cell 4 - Analysis of the data:

```python
# Calculate statistics
mean_price = np.mean(ticket_prices)
mean_tickets = np.mean(number_of_tickets)
correlation = np.corrcoef(number_of_tickets, ticket_prices)[0, 1]

print("=" * 50)
print("STATISTICAL ANALYSIS")
print("=" * 50)

print(f"\nDESCRIPTIVE STATISTICS:")
print(f"• Average ticket price: ${mean_price:.2f}")
print(f"• Average number of tickets: {mean_tickets:.1f}")
print(f"• Price range: ${min(ticket_prices):.2f} - ${max(ticket_prices):.2f}")
print(f"• Ticket range: {min(number_of_tickets)} - {max(number_of_tickets)}")

print(f"\nCORRELATION:")
print(f"• Correlation coefficient: {correlation:.3f}")
if abs(correlation) > 0.9:
    print("• This indicates a VERY STRONG positive correlation")
elif abs(correlation) > 0.7:
    print("• This indicates a STRONG positive correlation")
else:
    print("• This indicates a MODERATE positive correlation")

# Calculate price per ticket ratio
price_per_ticket = [p/n for p, n in zip(ticket_prices, number_of_tickets)]
avg_price_per_ticket = np.mean(price_per_ticket)

print(f"\nPRICE ANALYSIS:")
print(f"• Average price per ticket: ${avg_price_per_ticket:.3f}")
print("• Individual price per ticket:")
for i in range(len(number_of_tickets)):
    print(f"  {number_of_tickets[i]} tickets for ${ticket_prices[i]:.2f} = ${price_per_ticket[i]:.3f} per ticket")
```
