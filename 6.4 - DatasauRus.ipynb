# The Datasaurus

... is yet another dataset. It's a fun little dataset to compare the different graphs data with the same statiscal properties can have.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("files/datasauRus.csv")  
    
df.head(10) 
```

Explore the dataset-field. How many unique values are there?

```python

print("DATASAURUS DATASET")
print("=" * 70)
print("First 10 rows:")
print(df.head(10))
print()

# Explore the dataset field
print("UNIQUE VALUES IN 'dataset' COLUMN")
print("=" * 70)
unique_datasets = df['dataset'].unique()
print(f"Number of unique datasets: {len(unique_datasets)}")
print(f"Unique dataset names: {unique_datasets}")

# Count rows per dataset
print("\nROWS PER DATASET")
print("=" * 70)
dataset_counts = df['dataset'].value_counts()
print(dataset_counts)

# Basic statistics for each dataset
print("\nSTATISTICS BY DATASET")
print("=" * 70)
stats = df.groupby('dataset')[['x', 'y']].agg(['mean', 'std', 'min', 'max'])
print(stats.round(2))
```

First print the stats. Describe the dataframe grouped by the dataset-field.

```python


print("DATASAURUS DATASET - STATISTICS BY DATASET")
print("=" * 80)

# Group by dataset and calculate statistics
grouped_stats = df.groupby('dataset')[['x', 'y']].describe()

print(grouped_stats.round(2))
print()

# Alternative: Specific statistics only
print("\nSUMMARY STATISTICS (mean, std, min, max):")
print("=" * 80)
specific_stats = df.groupby('dataset')[['x', 'y']].agg(['mean', 'std', 'min', 'max'])
print(specific_stats.round(2))
print()

# Check if statistics are indeed identical across datasets
print("\nVERIFICATION - Are statistics identical across datasets?")
print("=" * 80)

# Check mean of x
x_means = df.groupby('dataset')['x'].mean()
print(f"X means are all equal: {np.allclose(x_means, x_means.iloc[0])}")

# Check mean of y
y_means = df.groupby('dataset')['y'].mean()
print(f"Y means are all equal: {np.allclose(y_means, y_means.iloc[0])}")

# Check std of x
x_stds = df.groupby('dataset')['x'].std()
print(f"X stds are all equal: {np.allclose(x_stds, x_stds.iloc[0])}")

# Check std of y
y_stds = df.groupby('dataset')['y'].std()
print(f"Y stds are all equal: {np.allclose(y_stds, y_stds.iloc[0])}")
```

Now create a pivot_table just showing the mean and the std for X and Y indexed on the dataset-field. Which means the dataset field appears in the rows of the pivot table.

```python

pivot_stats = pd.pivot_table(
    df,
    values=['x', 'y'],
    index='dataset',
    aggfunc=['mean', 'std']
)

print("PIVOT TABLE - MEAN AND STD FOR X AND Y BY DATASET")
print("=" * 80)
print(pivot_stats.round(2))
```

In the pivottable you cannot add the correlation between X and Y.  So now group the dataframe again on the dataset-field.  And calculate the correlation between X and Y per dataset.

```python


# Load the dataset
df = pd.read_csv("files/datasauRus.csv")

# Group by dataset and calculate correlation between X and Y
correlation_by_dataset = df.groupby('dataset')[['x', 'y']].corr().iloc[0::2, -1]

print("CORRELATION BETWEEN X AND Y BY DATASET")
print("=" * 80)
print(correlation_by_dataset.round(4))
print()

# Alternative: More readable format
corr_values = {}
for dataset in df['dataset'].unique():
    subset = df[df['dataset'] == dataset]
    corr = subset['x'].corr(subset['y'])
    corr_values[dataset] = corr

corr_df = pd.DataFrame.from_dict(corr_values, orient='index', columns=['correlation'])
print("CORRELATION VALUES (alternative method):")
print("=" * 80)
print(corr_df.round(4))
```

All numbers are pretty much the same. However, when you plot the individual observations for X and Y per datasets on a scatter, you will find out that they are completely different.

Each dataset leads to another figure. So now try to plot just the dino dataset on a scatter. You will immediately understand why this dataset is called 'dinosaurus'.

```python


# Load the dataset
df = pd.read_csv("files/datasauRus.csv")

# Filter for the 'dino' dataset
dino_data = df[df['dataset'] == 'dino']

# Create the scatter plot
plt.figure(figsize=(10, 8))
plt.scatter(dino_data['x'], dino_data['y'], color='darkgreen', alpha=0.7, s=30)
plt.title('Datasaurus - Dino Dataset', fontsize=16)
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)
plt.grid(True, alpha=0.3)
plt.axis('equal')  # Equal aspect ratio
plt.tight_layout()
plt.show()

# Print the statistics to confirm they match the others
print("DINO DATASET STATISTICS:")
print(f"X - Mean: {dino_data['x'].mean():.2f}, Std: {dino_data['x'].std():.2f}")
print(f"Y - Mean: {dino_data['y'].mean():.2f}, Std: {dino_data['y'].std():.2f}")
print(f"Correlation: {dino_data['x'].corr(dino_data['y']):.4f}")
```

You can now try to plot all the diffent datasets on the same canvas. The code looks like this.

```python
# plot groupby results on the same canvas 
fig, ax = plt.subplots(figsize=(8,6))

for n, grp in df.groupby('dataset'):
    ax.scatter(x = "x", y = "y", data=grp, label=n)
ax.legend(title="Label")

plt.show()
```

Not quite what we were after. Try to plot every group in a separate graph, all on the same canvas. There are 13 datasets.

* First use plt.subplots to create a canvas with 5 rows and 3 colums
* Next for-loop over the grouping of df on the dataset-field
* In every for-loop, add a scatterplot for the X and Y-data in the group.
    * Add it at the correct location. The following code is fun: <code>for i in range(15): print(i//3, i%3)</code>
    * You can get i by simply creating (and increasing a variable) or by enumerating the groupby
    * Try using <code>plt.cm.tab20</code> to give every graph another color!
* Delete the plots at positions 4,1 and 4,2 (13 is prime, after all)
* Show the plot

```python

# Create a canvas with 5 rows and 3 columns
fig, axes = plt.subplots(5, 3, figsize=(15, 20))

# Flatten axes for easy indexing
axes = axes.flatten()

# Get unique datasets
datasets = df['dataset'].unique()
print(f"Number of datasets: {len(datasets)}")

# Loop through datasets with enumeration
for i, dataset_name in enumerate(datasets):
    # Filter data for current dataset
    subset = df[df['dataset'] == dataset_name]
    
    # Get row and column position
    row = i // 3
    col = i % 3
    print(f"Dataset {i}: {dataset_name} → position ({row}, {col})")
    
    # Plot in the appropriate subplot
    axes[i].scatter(
        subset['x'], 
        subset['y'], 
        alpha=0.7, 
        s=15,
        color=plt.cm.tab20(i)  # Different color for each dataset
    )
    axes[i].set_title(dataset_name, fontsize=12)
    axes[i].set_xlim(0, 100)
    axes[i].set_ylim(0, 100)
    axes[i].set_aspect('equal')
    axes[i].grid(True, alpha=0.3)

# Delete plots at positions 4,1 and 4,2 (indices 12 and 13)
# Since we have 13 datasets, indices 0-12 are used
# We need to remove axes[13] and axes[14] (the extra subplots)
for j in range(13, 15):
    axes[j].set_visible(False)
    print(f"Hiding extra subplot at index {j}")

# Add main title
plt.suptitle('All Datasaurus Datasets - 13 Different Shapes, Same Statistics', 
             fontsize=16, y=0.98)

# Adjust layout
plt.tight_layout()
plt.show()
```
