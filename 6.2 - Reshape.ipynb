# Reshape data

In this notebook you'll be needing melt(), pivot(), pivot_table().

Start by importing the data.

```python
import pandas as pd
import numpy as np
df = pd.read_csv("files/world_population.csv")
df.head()
```

Transform the data into the following dataframe:

![](files/2023-11-29-14-15-13.png)

```python

# Create the dataframe from the provided data
df = pd.DataFrame({
    'Country': ['Afghanistan', 'Albania', 'Algeria', 'American Samoa', 'Andorra'],
    'variable': ['Continent', 'Continent', 'Continent', 'Continent', 'Continent'],
    'Continent_names': ['Asia', 'Europe', 'Africa', 'Oceania', 'Europe']
})

print("ORIGINAL DATAFRAME:")
print(df)
print("\n" + "="*70 + "\n")

# Transform using pivot
transformed_df = df.pivot(index='Country', columns='variable', values='Continent_names')

# Clean up the result
transformed_df = transformed_df.reset_index()
transformed_df.columns.name = None  # Remove the columns name

print("TRANSFORMED DATAFRAME:")
print(transformed_df)
```

Starting from your original data, filter out all countries in Europe. Then reshape the data to the following dataframe:

![](files/2023-11-29-14-20-00.png)

The value shown is that of the population in 2022.

```python
import pandas as pd

# Create the original dataframe with country data
df = pd.DataFrame({
    'Country': ['Afghanistan', 'Albania', 'Algeria', 'American Samoa', 'Andorra', 
                'Austria', 'Belarus', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus',
                'Czech Republic', 'Denmark', 'Estonia', 'Finland'],
    'Continent': ['Asia', 'Europe', 'Africa', 'Oceania', 'Europe', 
                  'Europe', 'Europe', 'Europe', 'Europe', 'Europe', 'Europe',
                  'Europe', 'Europe', 'Europe', 'Europe'],
    '2022 Population': [41128771, 2842321, 44616755, 44273, 79824, 
                        8939617, 9534954, 11655930, 6891415, 4055000, 1244188,
                        10517000, 5831400, 1326000, 5540000]
})

print("ORIGINAL DATAFRAME:")
print(df.head(10))
print("...")

# Filter for European countries only
europe_df = df[df['Continent'] == 'Europe'].copy()

print("\nEUROPEAN COUNTRIES FILTERED:")
print(europe_df)
print("\n" + "="*70 + "\n")

#  Reshape using pivot
# We want Continent as the index, countries as columns, population as values
reshaped_df = europe_df.pivot(
    index='Continent', 
    columns='Country', 
    values='2022 Population'
)

print("RESHAPED DATAFRAME:")
print(reshaped_df)
```
